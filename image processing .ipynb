```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_image(image_1, image_2, title_1="original", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot (1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
pic = cv2.imread("photo.jpg")

# Convert the color space from BGR to RGB (matplotlib uses RGB)
pic_rgb = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(pic_rgb)
plt.axis('off')  # Optional: Turn off axis
plt.show()
```

```python
def plot_hist(old_image, new_image,title_old="Orignal", title_new="New Image"):
    intensity_values = np.array([x for x in range(256)])
    plt.subplot(1, 2, 1)
    plt.bar(intensity_values, cv2.calcHist([old_image], [0],None, [256], [0,256])[:,0],width = 5)
    plt.title(title_old)
    plt.xlabel('intensity')
    plt.subplot(1, 2, 2)
    plt.bar(intensity_values, cv2.calcHist([new_image], [0],None, [256], [0,256]) [:,0],width = 5)
    plt.title(title_new)
    plt.xlabel('intensity')
    plt.show()
```

```python
goldhill = cv2.imread("photo.jpg",cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(5,5))
plt.imshow(goldhill,cmap="gray")
plt.show()
```

```python
hist = cv2.calcHist([goldhill], [0], None, [256], [0,256])
```

```python
intensity_values = np.array([x for x in range(hist.shape[0])])
plt.bar(intensity_values, hist[:,0], width = 5)
plt.title("Bar histogram")
plt.show()
```

```python
plt.plot(intensity_values,hist) 
plt.title("histogram")
plt.show()
```

```python
photo = cv2.imread("photo.jpg")

# Check if the image is loaded successfully
if photo is None:
    print("Error: Unable to load image.")
else:
    # Proceed with histogram calculation
    color = ('blue', 'green', 'red')
    for i, col in enumerate(color): 
        histr = cv2.calcHist([photo], [i], None, [256], [0,256])
        plt.plot(histr, color = col, label=col+" channel")
        plt.xlim([0,256])

    plt.legend()
    plt.title("Histogram Channels")
    plt.show()
```

```python
alpha=2       # Simple contrast control
beta=-50       # Simple brightness control
new_image = cv2.convertScaleAbs (goldhill, alpha=alpha, beta=beta)
plot_image(goldhill, new_image, title_1= "Ocignal", title_2 = "brightness, and contrast control")
```

```python
plt.figure(figsize=(10,5)) 
plot_hist(goldhill, new_image, "Orignal", "brightness control")
```
